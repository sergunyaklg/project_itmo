```python
import pandas as pd
data_from_baza = pd.read_csv('report.csv', delimiter=',')
data_from_baza
```

```python
train_data = data_from_baza[((data_from_baza['TARGET'] == 0) & (data_from_baza['MIP'] >= 94.6640625 ) & (data_from_baza['MIP'] <= 95.2890625)) | ((data_from_baza['TARGET'] == 1) & (data_from_baza['MIP'] >= 52.9296875) & (data_from_baza['MIP'] <= 58.7890625))]
train_data.sort_values(by=['MIP'], inplace=True)
train_data = train_data.reset_index()
train_data
```

```python
train_data.mean()
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler() 
scaler.fit(train_data) 
scaled = scaler.fit_transform(train_data) 
scaled_df = pd.DataFrame(scaled, columns=train_data.columns)
scaled_df
```

```python
scaled_df.mean()
```

```python
X = pd.DataFrame(scaled_df.drop(['index', 'TARGET'], axis=1))
X
```

```python
y = pd.DataFrame(scaled_df['TARGET'])
y
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression(random_state=2019, solver='lbfgs').fit(X, y.values.ravel())
```

```python
reg.predict_proba([[0.089, 0.122, 0.511, 0.237, 0.445, 0.513, 0.068, 0.857]])
```

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors = 3, p = 2)

neigh.fit(X, y)
```

```python
import numpy as np

NewObject = np.array([0.089, 0.122, 0.511, 0.237, 0.445, 0.513, 0.068, 0.857])
neigh.predict([NewObject])
```

```python
neigh.kneighbors([NewObject])
```
