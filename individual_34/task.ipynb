```python
import pandas as pd
from io import StringIO
```

```python
TESTDATA = StringIO("""Object,X,Y,Cluster
1, 3, 18, 0
2, 15, 18, 1
3, 15, 5, 2
4, 11, 8, 1
5, 5, 12, 2
6, 8, 7, 0
7, 12, 20, 2
8, 20, 1, 1
9, 9, 9, 0
10, 8, 19, 0
11, 13, 7, 0
12, 11, 16, 0
13, 15, 19, 2
14, 20, 7, 2
15, 16, 12, 1
    """)

df = pd.read_csv(TESTDATA, index_col = 'Object')
df
```

```python
import numpy as np
X = pd.DataFrame(df.drop(['Cluster'], axis=1))
centroid = np.array([[8.67, 12.67], [15.5, 9.75], [13.4, 12.6]])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init=centroid)
model = kmeans.fit(X)
model.labels_
```

```python
alldistances = model.transform(X)
alldistances
```

```python
print('Координаты центроидов: ', model.cluster_centers_)
print('Затрачено итераций: ', model.n_iter_)
```

```python
# Добавим номера назначенных кластеров
df['Assigned'] = model.labels_
# Добавим расстояние до центроида кластера 0
df['Dist_0'] = alldistances[:,0]
df
```

```python
df_0 = df[df['Assigned'] == 0]
df_0
```
